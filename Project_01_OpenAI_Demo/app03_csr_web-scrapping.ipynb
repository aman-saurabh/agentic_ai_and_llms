```python
from playwright.sync_api import sync_playwright, Playwright
from openai import OpenAI
import json

print("✓ Libraries imported successfully")
```

```python
OPENAI_API_KEY = "your-openai-api-key-here"
client = OpenAI(api_key=OPENAI_API_KEY)

print("✓ OpenAI client initialized")
```

```python
def scrape_website(url):
    """Scrape a client-side rendered website"""
    
    playwright = sync_playwright().start()
    browser = playwright.chromium.launch(headless=True)
    page = browser.new_page()
    
    # Navigate and wait for content to load
    page.goto(url, wait_until='networkidle', timeout=60000)
    page.wait_for_timeout(3000)  # Wait 3 seconds for JavaScript
    
    # Extract content
    title = page.title()
    text = page.evaluate('() => document.body.innerText')
    html = page.content()
    
    # Cleanup
    browser.close()
    playwright.stop()
    
    return {
        'url': url,
        'title': title,
        'text': text,
        'html': html
    }

# Scrape OpenAI website
url = "https://openai.com"
scraped_data = scrape_website(url)

print(f"✓ Successfully scraped: {scraped_data['title']}")
print(f"✓ Text content length: {len(scraped_data['text'])} characters")

# ============================================
# CELL 5: View Scraped Content
# ============================================
# Display first 500 characters of text
print("First 500 characters of scraped text:")
print("=" * 50)
print(scraped_data['text'][:500])
print("=" * 50)
```
